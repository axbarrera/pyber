```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
pyber_data_complete = pd.merge(ride_data, city_data, how = "right", on = ["city", "city"])
# Display the data table for preview
pyber_data_complete.head()


#testing writing to csv
#pyber_data_complete.to_csv('test.csv')

```

## Bubble Plot of Ride Sharing Data

```python
#Calculate average price by city for rural, urban and suburban areas

city_urban = (pyber_data_complete[pyber_data_complete['type'] == 'Urban'].groupby('city')).mean()
city_urban = pd.DataFrame(city_urban)
avg_fare_urban = city_urban["fare"]
#avg_fare_urban
rides_per_city_urban = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Urban'])
rides_per_city_urban = rides_per_city_urban["city"].value_counts()
#rides_per_city_urban


city_suburban = (pyber_data_complete[pyber_data_complete['type'] == 'Suburban'].groupby('city')).mean()
city_suburban = pd.DataFrame(city_suburban)
avg_fare_suburban = city_suburban["fare"]
#avg_fare_suburban
rides_per_city_suburban = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Suburban'])
rides_per_city_suburban = rides_per_city_suburban["city"].value_counts()
#rides_per_city_suburban


city_rural = (pyber_data_complete[pyber_data_complete['type'] == 'Rural'].groupby('city')).mean()
city_rural = pd.DataFrame(city_rural)
avg_fare_rural = city_rural["fare"]
#avg_fare_rural
rides_per_city_rural = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Rural'])
rides_per_city_rural = rides_per_city_rural["city"].value_counts()
#rides_per_city_rural


num_drivers_by_city = pyber_data_complete.groupby('city').mean()
num_drivers_by_city = num_drivers_by_city["driver_count"]
#num_drivers_by_city

#num_rides_per_city = pyber_data_complete[('city')].value_counts()
#num_rides_per_city



plt.scatter(rides_per_city_urban, avg_fare_urban, s=num_drivers_by_city*5,
            color = 'coral', alpha = 1, edgecolor= 'gray', label = 'Urban', linewidth=1)
plt.scatter(rides_per_city_suburban, avg_fare_suburban,s=num_drivers_by_city*5,
            color = 'skyblue', alpha = 1, edgecolor= 'gray', label = 'Suburban', linewidth=1)
plt.scatter(rides_per_city_rural, avg_fare_rural,s=num_drivers_by_city*5,
            color = 'gold', alpha = 1, edgecolor= 'gray', label = 'Rural', linewidth=1)
plt.title("Ride Sharing Metrics - Urban, Rural, Suburban")
plt.xlabel("Number of Rides per City")
plt.ylabel("Average Fare")
plt.show()
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
pyber_complete = pyber_data_complete.copy() 
total_fares = pyber_complete["fare"].sum()
#total_fares

rides_per_city_urban = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Urban'])
total_urban_fares = rides_per_city_urban["fare"].sum()
#total_urban_fares

rides_per_city_suburban = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Suburban'])
total_suburban_fares = rides_per_city_suburban["fare"].sum()
#total_suburban_fares

rides_per_city_rural = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Rural'])
total_rural_fares = rides_per_city_rural["fare"].sum()
#total_rural_fares

# Build Pie Chart
explode = (0.1, 0, 0)
colors = ["salmon", "skyblue", "yellow"]
labels = ["Urban", "Suburban", "Rural"]
type_totals_fares = [total_urban_fares,total_suburban_fares,total_rural_fares]
plt.pie(type_totals_fares, labels=labels, explode = explode, colors=colors, shadow = True, autopct="%1.1f%%")
plt.title("Total Fares by City Type")
plt.show()
# Save Figure
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
pyber_complete = pyber_data_complete.copy() 
total_rides = pyber_complete["ride_id"].count()
#total_rides

rides_per_city_urban = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Urban'])
rides_per_city_urban = rides_per_city_urban["ride_id"].count()
#rides_per_city_urban

rides_per_city_suburban = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Suburban'])
rides_per_city_suburban = rides_per_city_suburban["ride_id"].count()
#rides_per_city_suburban

rides_per_city_rural = pd.DataFrame(pyber_data_complete[pyber_data_complete['type'] == 'Rural'])
rides_per_city_rural = rides_per_city_rural["ride_id"].count()
rides_per_city_rural


explode = (0.1, 0, 0)
colors = ["salmon", "skyblue", "yellow"]
labels = ["Urban", "Suburban", "Rural"]
type_totals_rides = [rides_per_city_urban,rides_per_city_suburban,rides_per_city_rural]
plt.pie(type_totals_rides, labels=labels, explode = explode, colors=colors, shadow = True, autopct="%1.1f%%")
plt.title("Total Rides by City Type")
plt.show()

# Build Pie Chart

# Save Figure
```
